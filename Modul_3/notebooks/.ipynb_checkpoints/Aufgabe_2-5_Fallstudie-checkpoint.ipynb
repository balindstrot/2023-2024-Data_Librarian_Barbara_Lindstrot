```python
import pandas as pd
df = pd.read_csv("../data/Library_Usage.csv")
df.columns
```

```python
df['Circulation Active Year'] = pd.to_numeric(
     df['Circulation Active Year'], errors='coerce')
```

```python
df.head()
```

```python
df['Circulation Active Month'] = pd.to_datetime(
    df['Circulation Active Month'],
    errors='coerce',
    format="%b"
).dt.month
```

```python
df['Circulation Active Month'].dtype
```

```python
df['Circulation Active Year'].dtype
```

Berechnung Mitgliedschaft in Monaten:
'Membership Duration' = ('Circulation Active Year' - 'Year Patron Registered')*12 + 'Circulation Active Month'

```python
#for i in df['Circulation Active Year']:
Membership_Duration = (df['Circulation Active Year'] - df['Year Patron Registered'] ) * 12 + df['Circulation Active Month']
```

```python
df.insert(loc=9, column = 'Membership Duration in Month', value = Membership_Duration)
```

```python
df.head()
```

```python
Membership_Duration_year = ((df['Circulation Active Year'] - df['Year Patron Registered'] ) * 12 + df['Circulation Active Month'])/12
```

```python
df.insert(loc=10, column = 'Membership Duration in Years', value = Membership_Duration_year )
```
