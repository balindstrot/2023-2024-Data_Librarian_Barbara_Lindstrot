```python
import pyterrier as pt
import requests
import pandas as pd
import pickle
import nltk
import numpy as np

```

```python
url = f"https://www.bibsonomy.org/json/search/j%C3%BClich?items=1000"
```

```python
res = requests.get(url)
```

```python
data = res.json()
```

```python
data
```

```python
df_juelich = pd.DataFrame(data["items"])
#df_juelich = df_juelich[df_juelich["type"] == "Publication"]
```

```python
df_juelich = df_juelich.dropna(axis=1, thresh=len(df_juelich)*0.5)
```

```python
df_juelich.head(10)
```

```python
df_juelich.columns
```

```python
useful_columns = ['id', 'tags', 'label', 'description', 'authors', 'year']
df_juelich = df_juelich[useful_columns]
```

```python
if not pt.started():
    pt.init()
```

```python
df_juelich['docno'] = df_juelich['id']
df_juelich['text'] = df_juelich['label']
```

```python
index_folder ="./juelich_index"

indexer = pt.DFIndexer(index_folder, overwrite=True)
index_ref =   indexer.index(df_juelich['text'], df_juelich['docno'])
print(f"path to our index: {index_ref.toString()}")
```

```python
index = pt.IndexFactory.of(index_ref)
```

```python
print(index.getCollectionStatistics().toString())
```

```python
for kv in index.getLexicon():
    print(f"{kv.getKey()} -> {kv.getValue().toString()}")
```

```python
term = "forschungszentrum"
index.getLexicon()[term].toString()
```

```python
term_freq_dict = {}

for kv in index.getLexicon():
    term_freq_dict[kv.getKey()] = kv.getValue().frequency
```

```python
def get_idf_for_term(term, index):
    lex = index.getLexicon()
    stemmed_term = stemmer.stem(term)

    if not stemmed_term in lex:
        return

    lex_entry = lex[stemmed_term]
    
    df_term = lex_entry.getDocumentFrequency()
    N = index.getCollectionStatistics().numberOfDocuments
    
    idf = N/df_term
    
    log_idf = np.emath.logn(10, idf)
    
    return log_idf
```

```python
get_idf_for_term("Forschungszentrum", index)
```

```python
#define your own tf_idf_method here:
def calc_tf_idf(query, docno,index):
    lex = index.getLexicon()
    stemmed_term = stemmer.stem(query)

    if not stemmed_term in lex:
        return

    lex_entry = lex[stemmed_term]
    
    df_term = lex_entry.getDocumentFrequency()
    N = index.getCollectionStatistics().numberOfDocuments
    
    idf = N/df_term
    
    log_idf = np.emath.logn(10, idf)
    
    return log_idf
    #remember that tfidf is the product of two components
    #hint: the tf model seach result contains tf frequencies
    
```

```python
calc_tf_idf(query, docno,index)
```
