```python
import json
import requests
import pandas as pd
```

```python
url1 = "https://www.bibsonomy.org/json/search/Bibliothek?items=1000&duplicates=merged"
url2 = "https://www.bibsonomy.org/json/search/Data%20Librar*?items=1000&duplicates=merged"
```

```python
ergebnis1 = requests.get(url1)
ergebnis2 = requests.get(url2)
```

```python
daten1 = ergebnis1.json()
daten2 = ergebnis2.json()
```
